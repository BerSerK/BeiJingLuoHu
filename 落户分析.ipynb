```python
import pandas as pd
def read_data(filename):
    print(filename)
    raw_dfs = pd.read_html(filename)
    return raw_dfs[1]
dfs = []
for i in range(1, 62):
    dfs.append(read_data('data/落户公示%d.htm'%i))
df = pd.concat(dfs)
df['year'] = [int(a.split('-')[0]) for a in df['出生年月']]
df['age'] = 2019-df['year']
df.index = df['序号']
```

```python
df
```

```python
import matplotlib.pyplot as plt
data = df.groupby('year').size()
plt.bar(data.index, data)
plt.grid(True)
plt.xticks(rotation=90)
display(data)
```
