```python
import datetime
df['年龄'] = datetime.datetime(2020, 10, 15) - df['生日']
df['年龄'] = [d.days/365 for d in df['年龄']]
df['出生年份']  = [d.year for d in df['生日']]
```

```python
gp = df.groupby('公司')
danwei_df = pd.DataFrame({'人数':gp.size(),'平均年龄':gp['年龄'].mean()})
danwei_df = danwei_df.sort_values('人数', ascending = False)
danwei_df.iloc[:10]
```

```python
import pandas as pd
def read_data(filename):
    print(filename)
    raw_dfs = pd.read_html(filename)
    return raw_dfs[1]
dfs = []
for i in range(1, 62):
    dfs.append(read_data('data/落户公示%d.htm'%i))
df2019 = pd.concat(dfs)
df2019['year'] = [int(a.split('-')[0]) for a in df2019['出生年月']]
df2019['age'] = 2019-df['year']
df2019.index = df['序号']
```

```python
import matplotlib.pyplot as plt

data2019 = df2019.groupby('year').size()
plt.bar(data2019.index, data2019)
plt.grid(True)
plt.xticks(rotation=90)

data = df.groupby('出生年份').size()
plt.bar(data.index, data)
plt.grid(True)
plt.xticks(rotation=90)
display(data)
```
